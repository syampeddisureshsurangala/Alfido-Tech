# **Task - 1 Weather Data Analysis **

**step-1 Import Libraries **

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Step-2 Load the Data **

```python
file_path = '/content/Weather Data.csv'
```

```python
df = pd.read_csv(file_path)
```

**Step-3 Explore the Data **

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

** Step-4 Data Cleaning**

```python
df.isnull().sum()
df = df.dropna()
```

**Step-5 Data Analysis**

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Weather', data=df)
plt.title('Weather conditions')
plt.xlabel('Weather')
plt.ylabel('Count')
plt.xticks(rotation=90) # Rotate x-axis labels by 90 degrees
plt.show()
```

```python
sns.histplot(df['Rel Hum_%'], bins=20, kde=True)
plt.title('Distribution of Relative Humidity')
plt.xlabel('Relative Humidity (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(x=df["Wind Speed_km/h"], bins=20, kde=True)
plt.title('Wind Speed')
plt.xlabel('Wind Speed (km/h)')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Temp_C'], df['Rel Hum_%'], alpha=0.5)
plt.title('Temperature vs. Relative Humidity')
plt.xlabel('Temperature (C)')
plt.ylabel('Relative Humidity (%)')
plt.grid(True)
plt.show()
```

```python
sns.pairplot(df[['Date/Time', 'Temp_C', 'Dew Point Temp_C', 'Rel Hum_%',
       'Press_kPa', 'Weather', 'Visibility_km']])
plt.show()
```

```python
# analayze temparature trends

df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df.set_index('Date/Time', inplace=True)

df['Temp_C'].plot(figsize=(12,6))
plt.title('Temperature Trends')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (C)')
plt.show()
```
