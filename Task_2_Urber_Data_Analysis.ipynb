**Urber Data Analysis**

Import Necessary Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the Data

```python
filepath='/content/UberDataset.csv'
df = pd.read_csv(filepath)
```

 Explore the Data

 Data Cleaning


```python
df.isnull().sum()
df.dropna(inplace=True)
```

Data Analysis

```python
#Analyze ride times to understand peak hours

# Convert 'START_DATE' column to datetime with the appropriate format
df['START_DATE'] = pd.to_datetime(df['START_DATE'], format='%m-%d-%Y %H:%M', errors = 'coerce')

# Extract hour of the day from 'START_DATE'
df['HOUR_OF_DAY'] = df['START_DATE'].dt.hour

# Now, let's proceed with the analysis
plt.figure(figsize=(10, 6))
sns.boxplot(x='HOUR_OF_DAY', y='MILES', data=df)
plt.title('Ride Distances by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Ride Distance (miles)')
plt.show()
```

```python
#Visualize pickup and drop-off locations:
plt.figure(figsize=(8, 10))
sns.scatterplot(x='START', y='STOP', data=df, hue='CATEGORY', palette='Set1', s=100)
plt.title('Start and Stop Locations of Rides')
plt.xticks([])
plt.yticks([])
plt.xlabel('Start')
plt.ylabel('Stop')
plt.legend(title='Category', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
#Analyze user ratings to understand satisfaction levels

plt.figure(figsize=(10, 6))
sns.boxplot(x='CATEGORY', y='MILES', data=df)
plt.title('Ride Distances by Category')
plt.xlabel('Category')
plt.ylabel('Ride Distance (miles)')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='PURPOSE', y='MILES', data=df)
plt.title('Miles by Purpose')
plt.xlabel('Purpose')
plt.ylabel('Miles')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='PURPOSE', data=df, palette='Set1')
plt.title('Distribution of Ride Purposes')
plt.xlabel('Purpose')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['MILES'], bins=20, kde=True)
plt.title('Distribution of Ride Distances')
plt.xlabel('Distance (miles)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.countplot(x='CATEGORY', data=df, palette='Set1')
plt.title('Distribution of Ride Categories')
plt.xlabel('Category')
plt.ylabel('Number of Rides')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.pie(df['CATEGORY'].value_counts(), labels=df['CATEGORY'].unique(), autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Ride Categories')
plt.show()
```
